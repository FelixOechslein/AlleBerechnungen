```python
import pandas as pd
df_transactions = pd.read_csv("./transactions_thalia.csv")
```

```python
df_t = df_transactions
df_t.head()
```

```python
df_t2 = df_t[["iid", "key", "attrtype", "status", "pricetotal"]]
df_t2.head()
```

```python
df_t2_attrtype_bereinigt = df_t2[df_t2["attrtype"] == "TI"]
```

```python
df_t2_attrtype_status_bereinigt = df_t2_attrtype_bereinigt[df_t2_attrtype_bereinigt["status"] == ("Buchung" or "Order")]
```

```python
list_t2_grouped = df_t2_attrtype_status_bereinigt.groupby(["iid"])["pricetotal"].agg("sum")
```

```python
list_t2_grouped.head()
```

```python
df_t2_grouped = list_t2_grouped.to_frame()
```

```python
export_csv = df_t2_grouped.to_csv ('df_CV_bereinigt_talia.csv')
```
