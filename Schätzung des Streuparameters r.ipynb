```python
# Annahme: df_trans_werte ist der DataFrame, der die Spalte "1" enthält

# Extrahiere die Spalte "1" als abhängige Variable 'y'
y = df_trans_werte['2'].values

def zinb_pmf(y, pi, r, p):
    nb_pmf = stats.nbinom.pmf(y, r, p)
    zi_pmf = np.where(y == 0, pi, 1 - pi)
    return nb_pmf * zi_pmf

def zinb_log_likelihood(params, y):
    pi, r, p = params
    loglik = np.sum(np.log(zinb_pmf(y, pi, r, p)))
    return -loglik

initial_params = np.array([0.5, 1, 0.5])

result = opt.minimize(zinb_log_likelihood, initial_params, args=(y,))
estimated_params = result.x

print("Estimated Parameters:")
print("pi =", estimated_params[0])
print("r =", estimated_params[1])
print("p =", estimated_params[2])
```
