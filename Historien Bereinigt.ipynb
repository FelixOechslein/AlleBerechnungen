```python
df_history = pd.read_csv("./history_thalia.csv")
```

```python
df_h = df_history.groupby(['iid'])['attrtype'].value_counts()
```

```python
df_h.to_frame()
```

```python
df_h_unstacked = df_h.unstack()
```

```python
df_h_unstacked.head()
```

```python
df_h_unstacked.info()
```

```python
df_keineOeffnungen_talia = df_h_unstacked[numpy.isnan(df_h_unstacked["NO"]) & numpy.isnan(df_h_unstacked["NC"])]
```

```python
export_csv = df_keineOeffnungen_talia.to_csv ('df_keineOeffnungen_talia.csv')
```
